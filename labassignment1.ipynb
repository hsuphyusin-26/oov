```python


import re
pattern=r'^[\u1000-\u109F\s]+$'
texts=[
 "ပင်မ",
 "မင်တံ123",
 "နေကောင်းလား",
 "ကျောင်းသား 1"
]
for t in texts:
  if re.match(pattern,t):
    print(f"Match:{t}")
  else:
    print(f"No match:{t}")
```

```python
import re
import random
rules=[
    (r".*i[']?m(depressed|sad).*",
     "I AM SORRY TO HEAR TOU ARE \\1"),
    (r".*i am(depressed|sad).*",
     "WHY DO YOU THINK YOU ARE \\1"),
    (r".*all.*",
     "IN WHAT WAY"),
    (r".*always.*",
     "CAN YOU THINK OF A SPECIFIC EXAMPLES"),
    (r".*my(.*)",
     "YOUR\\1"),
     (r".*",
      "PLEASE TELL ME MORE")
]
def eliza_response(user_input):
  user_input=user_input.lower()
  for pattern,response in rules:
    match=re.match(pattern,user_input)
    if match:
      return re.sub(pattern,response,user_input).upper()

print("ELIZA:HELLO.HOW ARE YOU FEEELING TODAY?")
while True:
      user=input("YOU:")
      if user.lower()in ["bye","exit","quit"]:
        print("ELIZA:GOODBYE.TAKE CARE.")
        break
      print("ELIZA:",eliza_response(user))
```
